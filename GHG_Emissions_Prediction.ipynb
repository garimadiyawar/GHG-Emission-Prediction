# Step 2: Load Dataset

```python
excel_file = 'SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx'  # Replace with actual path
years = range(2010, 2017)
```

```python
years[2]
```

```python
df_1 = pd.read_excel(excel_file, sheet_name=f'{years[0]}_Detail_Commodity')
df_1.head()
```

```python
df_2 = pd.read_excel(excel_file, sheet_name=f'{years[0]}_Detail_Industry')
df_2.head()
```

```python
all_data = []  # List to store combined data for all years

for year in years:
    try:
        # Read data from Excel sheets for Commodity and Industry for the given year
        df_com = pd.read_excel(excel_file, sheet_name=f'{year}_Detail_Commodity')
        df_ind = pd.read_excel(excel_file, sheet_name=f'{year}_Detail_Industry')

        # Add source and year information to each DataFrame
        df_com['Source'] = 'Commodity'
        df_ind['Source'] = 'Industry'
        df_com['Year'] = year
        df_ind['Year'] = year

        # Strip any leading/trailing whitespace from column names
        df_com.columns = df_com.columns.str.strip()
        df_ind.columns = df_ind.columns.str.strip()

        # Rename columns for consistency across both datasets
        df_com.rename(columns={
            'Commodity Code': 'Code',
            'Commodity Name': 'Name'
        }, inplace=True)

        df_ind.rename(columns={
            'Industry Code': 'Code',
            'Industry Name': 'Name'
        }, inplace=True)

        # Combine the two DataFrames and append to the master list
        combined_df = pd.concat([df_com, df_ind], ignore_index=True)
        all_data.append(combined_df)

    except Exception as e:
        # Print an error message if any issue occurs while processing a year's data
        print(f"Error processing year {year}: {e}")
```
